Question 1 : Explain the fundamental differences between DDL, DML, and DQL
commands in SQL. Provide one example for each type of command.

Ans: In SQL, commands are grouped based on what they do with the database. The three important categories are DDL, DML, and DQL:

🔹 1. DDL (Data Definition Language)

Purpose: Used to define or modify the structure of the database (tables, schemas, etc.)

Effect: Changes the database structure permanently

Common commands: CREATE, ALTER, DROP, TRUNCATE

Example:

CREATE TABLE Employees (
    ID INT,
    Name VARCHAR(50),
    Salary DECIMAL(10,2)
);

👉 This creates a new table named Employees.

🔹 2. DML (Data Manipulation Language)

Purpose: Used to insert, update, or delete data in tables

Effect: Modifies the data but not the structure

Common commands: INSERT, UPDATE, DELETE

Example:

INSERT INTO Employees (ID, Name, Salary)
VALUES (1, 'John', 50000);

👉 This adds a new record to the Employees table.

🔹 3. DQL (Data Query Language)

Purpose: Used to retrieve data from the database

Effect: Does not modify data, only fetches it

Main command: SELECT

Example:

SELECT * FROM Employees;

👉 This retrieves all records from the Employees table.

✅ Key Differences Summary:
Feature	DDL	DML	DQL
Purpose	Define structure	Modify data	Retrieve data
Affects	Database schema	Table data	Data only (read)
Example	CREATE, ALTER	INSERT, UPDATE, DELETE	SELECT

    

Question 2 : What is the purpose of SQL constraints? Name and describe three common types
of constraints, providing a simple scenario where each would be useful.

Ans: SQL constraints are rules applied to table columns to ensure data accuracy, consistency, and integrity in a database. They prevent invalid data from being entered and enforce business rules.

🔹 1. NOT NULL Constraint

Purpose: Ensures that a column cannot have a NULL (empty) value.

Use Case: When a field is mandatory.

Example Scenario:
In an employee database, every employee must have a name.

CREATE TABLE Employees (
    ID INT,
    Name VARCHAR(50) NOT NULL
);

👉 This ensures that no employee record can be created without a name.

🔹 2. PRIMARY KEY Constraint

Purpose: Uniquely identifies each record in a table. It cannot contain NULL values and must be unique.

Use Case: To uniquely identify rows.

Example Scenario:
Each employee must have a unique ID.

CREATE TABLE Employees (
    ID INT PRIMARY KEY,
    Name VARCHAR(50)
);

👉 No two employees can have the same ID.

🔹 3. FOREIGN KEY Constraint

Purpose: Maintains a relationship between two tables by linking a column to the primary key of another table.

Use Case: To ensure referential integrity.

Example Scenario:
Each employee belongs to a department.

CREATE TABLE Departments (
    DeptID INT PRIMARY KEY,
    DeptName VARCHAR(50)
);

CREATE TABLE Employees (
    ID INT PRIMARY KEY,
    Name VARCHAR(50),
    DeptID INT,
    FOREIGN KEY (DeptID) REFERENCES Departments(DeptID)
);

👉 Ensures that an employee can only be assigned to an existing department.

🎯 Summary

NOT NULL → Prevents empty values

PRIMARY KEY → Ensures uniqueness

FOREIGN KEY → Maintains relationships between tables

 Q.3-Explain the difference between LIMIT and OFFSET clauses in SQL. How
would you use them together to retrieve the third page of results, assuming each page has 10 records?

Ans: Difference between LIMIT and OFFSET in SQL

Both clauses are used for pagination (retrieving data in parts):

🔹 LIMIT

Purpose: Specifies the maximum number of records to return

Example:

SELECT * FROM Employees
LIMIT 10;

👉 Returns only the first 10 records

🔹 OFFSET

Purpose: Skips a specified number of records before starting to return rows

Example:

SELECT * FROM Employees
OFFSET 10;

👉 Skips the first 10 records and returns the rest

🔹 Using LIMIT and OFFSET Together

LIMIT → how many records to fetch

OFFSET → how many records to skip

📄 Retrieve the 3rd page (10 records per page)

Step-by-step:

Page size = 10

Page number = 3

OFFSET = (page number - 1) × page size

OFFSET = (3 − 1) × 10 = 20

✅ Final Query:
SELECT * FROM Employees
LIMIT 10 OFFSET 20;

👉 This will:

Skip the first 20 records

Return the next 10 records (i.e., records 21 to 30)

🎯 Simple Understanding:

Page 1 → OFFSET 0

Page 2 → OFFSET 10

Page 3 → OFFSET 20

Question 4 : What is a Common Table Expression (CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage.

Ans: What is a Common Table Expression (CTE)?

A Common Table Expression (CTE) is a temporary result set in SQL that you can define using the WITH clause and then reference within a query.
It exists only for the duration of that query.

🔹 Syntax:
WITH cte_name AS (
    SELECT column1, column2
    FROM table_name
    WHERE condition
)
SELECT * FROM cte_name;
🌟 Main Benefits of CTE

Improves Readability

Breaks complex queries into simpler, logical parts

Reusability

You can reference the CTE multiple times within the same query

Simplifies Complex Queries

Useful for handling joins, aggregations, and subqueries

Supports Recursion

Can be used for hierarchical data (e.g., employee-manager relationships)

🔹 Simple Example

Scenario: Retrieve employees who earn more than the average salary.

WITH AvgSalary AS (
    SELECT AVG(Salary) AS AvgSal
    FROM Employees
)
SELECT Name, Salary
FROM Employees, AvgSalary
WHERE Employees.Salary > AvgSalary.AvgSal;

👉 Explanation:

The CTE AvgSalary calculates the average salary

The main query selects employees earning more than that average

🎯 In Short:

CTE = Temporary named result set

Defined using WITH

Makes queries cleaner and easier to manage

Question 5 : Describe the concept of SQL Normalization and its primary goals. Briefly
explain the first three normal forms (1NF, 2NF, 3NF).

Ans:Concept of SQL Normalization

Normalization is the process of organizing data in a database to reduce redundancy (duplicate data) and improve data integrity.

🎯 Primary Goals of Normalization

Eliminate duplicate data

Ensure data consistency

Improve data organization

Make database easier to maintain

🔹 First Normal Form (1NF)

Rule:

Each column must contain atomic (indivisible) values

No repeating groups or multiple values in a single column

Example Scenario:
❌ Not in 1NF:

ID	Name	Phone Numbers
1	John	12345, 67890

✅ In 1NF:

ID	Name	Phone Number
1	John	12345
1	John	67890
🔹 Second Normal Form (2NF)

Rule:

Must be in 1NF

All non-key attributes must depend on the entire primary key (no partial dependency)

Example Scenario:
If a table has a composite key (StudentID + CourseID):

❌ Not in 2NF:

StudentID	CourseID	StudentName
(StudentName depends only on StudentID, not full key)

✅ In 2NF:

Student Table: (StudentID, StudentName)

Course Table: (CourseID, CourseName)

Enrollment Table: (StudentID, CourseID)

🔹 Third Normal Form (3NF)

Rule:

Must be in 2NF

No transitive dependency (non-key attributes should not depend on other non-key attributes)

Example Scenario:
❌ Not in 3NF:
| EmpID | DeptID | DeptName |
(DeptName depends on DeptID, not directly on EmpID)

✅ In 3NF:

Employee Table: (EmpID, DeptID)

Department Table: (DeptID, DeptName)

🎯 Summary Table
Normal Form	Key Rule
1NF	Atomic values, no repeating groups
2NF	No partial dependency
3NF	No transitive dependency

Question 7 : Generate a report showing CustomerName, Email, and the
TotalNumberofOrders for each customer. Include customers who have not placed
any orders, in which case their TotalNumberofOrders should be 0. Order the results by CustomerName.

Ans:SQL Query
SELECT
    c.CustomerName,
    c.Email,
    COUNT(o.OrderID) AS TotalNumberOfOrders
FROM Customers c
LEFT JOIN Orders o
    ON c.CustomerID = o.CustomerID
GROUP BY
    c.CustomerID, c.CustomerName, c.Email
ORDER BY
    c.CustomerName;
🔍 Explanation

LEFT JOIN → Includes all customers, even if they have no orders

COUNT(o.OrderID) → Counts only matching orders

If no orders → returns 0 automatically

GROUP BY → Groups data per customer

ORDER BY → Sorts alphabetically by CustomerName

Question 9 : Write a SQL query that uses a Common Table Expression (CTE) and a
Window Function (specifically ROW_NUMBER() or RANK()) to display the
CategoryName, ProductName, and Price for the top 2 most expensive products in
each CategoryName.

Ans: SQL Query
WITH RankedProducts AS (
    SELECT
        c.CategoryName,
        p.ProductName,
        p.Price,
        ROW_NUMBER() OVER (
            PARTITION BY c.CategoryName
            ORDER BY p.Price DESC
        ) AS RankNum
    FROM Products p
    JOIN Categories c
        ON p.CategoryID = c.CategoryID
)

SELECT
    CategoryName,
    ProductName,
    Price
FROM RankedProducts
WHERE RankNum <= 2
ORDER BY
    CategoryName,
    Price DESC;
🔍 Explanation
🔹 CTE (RankedProducts)

Combines Products and Categories

Assigns a rank to each product within its category

🔹 Window Function
ROW_NUMBER() OVER (PARTITION BY c.CategoryName ORDER BY p.Price DESC)

PARTITION BY → Resets ranking for each category

ORDER BY Price DESC → Highest price gets rank 1

🔹 Final Filter
WHERE RankNum <= 2

👉 Keeps only top 2 expensive products per category

Apparel	T-Shirt Casual	20.00

Question 10 : You are hired as a data analyst by Sakila Video Rentals, a global movie
rental company. The management team is looking to improve decision-making by
analyzing existing customer, rental, and inventory data.
Using the Sakila database, answer the following business questions to support key strategic
initiatives.
Tasks & Questions:
1. Identify the top 5 customers based on the total amount they’ve spent. Include customer
name, email, and total amount spent.
2. Which 3 movie categories have the highest rental counts? Display the category name
and number of times movies from that category were rented.
3. Calculate how many films are available at each store and how many of those have
never been rented.
4. Show the total revenue per month for the year 2023 to analyze business seasonality.
5. Identify customers who have rented more than 10 times in the last 6 months.

Ans: 1. Top 5 Customers by Total Spending
SELECT
    c.first_name,
    c.last_name,
    c.email,
    SUM(p.amount) AS total_spent
FROM customer c
JOIN payment p
    ON c.customer_id = p.customer_id
GROUP BY
    c.customer_id, c.first_name, c.last_name, c.email
ORDER BY
    total_spent DESC
LIMIT 5;
✅ 2. Top 3 Movie Categories by Rental Count
SELECT
    cat.name AS category_name,
    COUNT(r.rental_id) AS rental_count
FROM category cat
JOIN film_category fc
    ON cat.category_id = fc.category_id
JOIN inventory i
    ON fc.film_id = i.film_id
JOIN rental r
    ON i.inventory_id = r.inventory_id
GROUP BY
    cat.name
ORDER BY
    rental_count DESC
LIMIT 3;
✅ 3. Films Available per Store & Never Rented
SELECT
    s.store_id,
    COUNT(i.inventory_id) AS total_films,
    SUM(CASE
            WHEN r.rental_id IS NULL THEN 1
            ELSE 0
        END) AS never_rented
FROM store s
JOIN inventory i
    ON s.store_id = i.store_id
LEFT JOIN rental r
    ON i.inventory_id = r.inventory_id
GROUP BY
    s.store_id;
✅ 4. Total Revenue per Month (Year 2023)
SELECT
    DATE_FORMAT(p.payment_date, '%Y-%m') AS month,
    SUM(p.amount) AS total_revenue
FROM payment p
WHERE YEAR(p.payment_date) = 2023
GROUP BY
    month
ORDER BY
    month;
✅ 5. Customers with More Than 10 Rentals (Last 6 Months)
SELECT
    c.first_name,
    c.last_name,
    c.email,
    COUNT(r.rental_id) AS total_rentals
FROM customer c
JOIN rental r
    ON c.customer_id = r.customer_id
WHERE r.rental_date >= DATE_SUB(CURDATE(), INTERVAL 6 MONTH)
GROUP BY
    c.customer_id, c.first_name, c.last_name, c.email
HAVING
    total_rentals > 10
ORDER BY
    total_rentals DESC;
🎯 Key Concepts Used

JOINs → combine multiple tables

GROUP BY → aggregation

SUM / COUNT → calculations

LEFT JOIN → detect missing data (never rented)

DATE functions → time-based analysis

HAVING → filter aggregated results
